# **TASK 1** - Exploratory Data Analysis

Q.1- **What is the distribution of bankruptcy and non-bankruptcy classes in the dataset? Are the classes balanced or imbalanced?**

```python
class_distribution = data['Bankrupt?'].value_counts()
print("Class distribution:\n", class_distribution)
```

```python
# Check if the classes are balanced or imbalanced
if class_distribution[0] == class_distribution[1]:
    print("The classes are balanced.")
else:
    print("The classes are imbalanced.")
```

Summarizing your analysis and observation

Analyzing the target variable (Bankrupt?) and counting the number of instances for each class (bankrupt and non-bankrupt).
Then calculating the ratio of bankruptcy and non_bankruptcy instances.
Then as the ratio of instances between the minority and majority classes is significantly skewed, then the class distribution is considered imbalance.

Q.2- **How does the distribution of the "Operating Profit Rate" differ between bankrupt and non-bankrupt companies? Can you create a suitable plot to visualize this difference?**

```python
import matplotlib.pyplot as plt
```

```python
data.isna().sum()
```

```python
data[' Operating Profit Rate'].isnull()
```

```python
bankrupt_data=data[data['Bankrupt?']==1]
non_bankrupt_data=data[data['Bankrupt?']==0]
bankrupt_op_rate = bankrupt_data[' Operating Profit Rate']
non_bankrupt_op_rate = non_bankrupt_data[' Operating Profit Rate']
plt.boxplot([bankrupt_op_rate, non_bankrupt_op_rate])
plt.xticks([1, 2], ['Bankrupt', 'Non-Bankrupt'])
plt.xlabel('Company Status')
plt.ylabel('Operating Profit Rate')
plt.title('Distribution of Operating Profit Rate')
plt.show()
```

Summarizing your analysis and observations

Firstly, filtering and separating the Operating Profit Rate for bankrupt and non_bankrupt instances.
Then on comparing both instances plot the graph.
Here in this graph above allows you to compare the central tendency, spread, and skewness of the "Operating Profit Rate" distribution between bankrupt and non-bankrupt companies.

Q.3- **plot a bar graph to show how many companies are bankrupt or not(already asked in first ques)**

```python
bankrupt_count = data[data['Bankrupt?'] == 1].shape[0]
non_bankrupt_count = data[data['Bankrupt?'] == 0].shape[0]
labels = ['Bankrupt', 'Non-Bankrupt']
counts = [bankrupt_count, non_bankrupt_count]
plt.bar(labels, counts)
plt.xlabel('Company Status')
plt.ylabel('Count')
plt.title('Distribution of Bankrupt and Non-Bankrupt Companies')
plt.show()
```

Summarizing your analysis and observation

Here from the above bar graph, we can observe that number of Non-bankrupt companies are more than bankrupt companies.

Q.5- **Plot a heatmap without using the bankrupt column(using the seaborn lib)**

```python
columns = [col for col in data.columns if col != 'Bankrupt?']
subset_data = data[columns]
correlation_matrix = subset_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
